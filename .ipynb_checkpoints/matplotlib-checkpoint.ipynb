# Matplotlib

## A simple example

````python
import matplotlib.pyplot as plt
import numpy as np

# Creating a plot
fig, ax = plt.subplots() # Create a figure containing a single axes.
ax.plot([1, 2, 3, 4], [2, 4, 6, 8]) # Plot some data on the axes.
````

````python
# Simpler method
plt.plot([1, 2, 3, 4], [2, 4, 6 ,8])
````

## Parts of a figure

<img title="a title" alt="Anatomy of a figure" src="https://matplotlib.org/stable/_images/anatomy.png">

````python
fig = plt.figure() # an empty figure with no Axes

fig, ax = plt.subplots() # a figure with a single axes
ax.set_title("title") # title
ax.set_xlim(-10) # x limit 
ax.set_ylim(10) # y limit
ax.set_xlabel("xtest") # x label
ax.set_ylabel("ytest") # y label

fig, axs = plt.subplots(2, 2) # a figure with a 2x2 grid of Axes
axs[0][1].set_title("title") # title with multiple axes
````

## The object-oriented interface and the pyplot interface

Two methods:
- Explicitly create figures and axes, and call methods on them (the "object-oriented (OO) style"
- Rely on pyplot to automatically create and manage the figures and axes, and use pyplot functions for plotting

````python
# OO

x = np.linspace(0, 2, 100)

# Note that even in the OO-style, we use '.pyplot.figure' to create the figure.
fig, ax = plt.subplots() # Create a figure and an axes.
ax.plot(x, x, label='linear') # Plot some data on the axes
ax.plot(x, x**2, label='quadratic') # Plot some more data one the axes
ax.plot(x, x**3, label='cubic') # ... and some more.
ax.set_xlabel('x label') # Add an x-label to the axes.
ax.set_ylabel('y label') # Add a y-label to the axes.
ax.set_title('Simple plot') # Add a title to the axes.
ax.legend() # Add a legend
````

````python
# pyplot 
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear') # Plot some data on the (implicit axes.)
plt.plot(x, x**2, label='quadratic') # etc.
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Simple plot')
plt.legend()
````

**If you want to use matplotlib in GUI applications consider: https://matplotlib.org/stable/gallery/index.html#user-interfaces**

In older examples, you may find examples that instead use the so-called pylab interface, via `from pylab import *`. This is strongly discouraged nowadays and deprecated.

## Plot functions

````python
def my_plotter(ax, data1, data2, param_dict):
    """
    A helper function to make a graph
    
    Parameters
    ----------
    ax: Axes
        The axes to draw to
    
    data1: array
        The x data
    
    data2: array
        The y data
    
    param_dict: dict
        Dictionary of kwargs to pass to ax.plot
    
    Returns
    -------
    out: list 
        list of artists added
    """
    out = ax.plot(data1, data2, **param_dict)
    return out
````

````python
data1, data2, data3, data4 = np.random.randn(4, 100)
print(np.random.randn(4, 3))
fix, ax = plt.subplots(1, 1)
my_plotter(ax, data1, data2, {'marker': 'x'})
````

````python
fix, (ax1, ax2) = plt.subplots(1, 2)
my_plotter(ax1, data1, data2, {'marker': 'x'})
my_plotter(ax2, data3, data4, {'marker': 'o'})
````
